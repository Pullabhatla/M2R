# Nearest neighbour search

```python
from Req import Map, nearest_neighbour, best_nn, gen_poly, gen_ran, held_karp
from Req.greedy import repeated_2_opt
from time import time
import matplotlib.pyplot as plt
```

## Roots of unity

We investigate the regular polygons generated by the roots of unity in the complex plane and represent our results as a Hamiltonian cycle in 2D space. The below program runs much faster than its brute force counterpart. It also produces an optimal tour due to the symmetries of the polygons.

```python
for i in range(3, 9):
    t0 = time()
    dummy = nearest_neighbour(gen_poly(i))
    t1 = time()
    t = t1 - t0
    print(f"Time elapsed: {round(t, 3)} seconds")
    dummy.show2d()
```
