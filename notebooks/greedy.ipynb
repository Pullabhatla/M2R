# Greedy tour

```python
from math import sin, cos, pi
from Req import Map, minimal_tour, greedy
from time import time
import matplotlib.pyplot as plt
from random import random
```

## Roots of unity

We investigate the regular polygons generated by the roots of unity in the complex plane and represent our results as a Map in 2D space.

```python
def gen_poly(n, dist_func=None):
    if dist_func:
        return Map([(cos((2*pi*i)/n), sin((2*pi*i)/n)) for i in range(n)], dist_func)
    else:
        return Map([(cos((2*pi*i)/n), sin((2*pi*i)/n)) for i in range(n)])
```

The below program runs much faster than its brute force counterpart. It also produces an optimal tour due to the symmetries of the polygons. 

```python
times = []
for i in range(3, 9):
    t0 = time()
    dummy = greedy(gen_poly(i))
    t1 = time()
    t = t1 - t0
    times.append(t)
    print(f"Time elapsed: {round(t, 3)} seconds")
    dummy.show2d()
```

## Random maps

A lot of the below maps are far from optimal

```python
def gen_ran(n, dist_func=None):
    if dist_func:
        return Map([(random(), random()) for i in range(n)], dist_func)
    else:
        return Map([(random(), random()) for i in range(n)])
```

```python
for i in range(3, 20):
    greedy(gen_ran(i)).show2d()
```

What if we considered all starting points for a particular example? We would see that some tours are closer to optimal than others.

```python
map = gen_ran(10)

for i in range(10):
    greedy(map, i).show2d()
```
