GRAPH CLASS TAKES IN EDGES AND WEIGHTS

```python
from GraphTheory.read import Graph
```

```python
D = Graph([(1,2), (2,3), (3,4)], [1,2,3])  # edges and the corresponding weights
```

```python
D.draw()
```

```python
D.draw(edge_weights=True)
```

```python
print(D.nodes())
print(D.number_nodes())
print(D.edges())
print(D.distances())
print(D.weighted_edges())
print(D.edge_dictionary())
```

TRY GRAPH WITH MULIEDGES

```python
D = Graph([(1,2), (2,3), (3,4), (2,1)], [1,2,3,2]) 
print(D.nodes())
print(D.number_nodes())
print(D.edges())
print(D.distances())
print(D.weighted_edges())
print(D.edge_dictionary())
```

```python
D.draw(edge_weights=True, multi_edges=True)  # one down side only show the most weighted edge in the multi edges bit
```

```python
D.draw(multi_edges=True)
```

```python
D.edge_dictionary().keys()
```

```python
D.edge_dictionary()['(1, 2)']  # easy access to edge weights if needed
```

```python
D.adjacency()
```

CONVERTING COST ADJACENCY MATRIX INTO GRAPH OBJECT

```python
from GraphTheory.read import GraphMatrix
import numpy as np 
```

```python
A = np.array([[0, 1, 2], [3, 0, 4], [5, 8, 0]])
```

```python
graph_A = GraphMatrix(A).graph()
```

```python
graph_A.draw()
```

```python
graph_A.draw(multi_edges=True)
```

```python
graph_A.neighbors(3)  # neighbors of node 3
```

```python
graph_A.distance_neighbors(3)  # distance of the neighbours of nodes 3 
```

```python
print(graph_A.radius(3, 8))  # neighbours of distance 8 away from node 3
print(graph_A.radius(3, 5))  # neighbours of distance 5 away from node 3
```

```python
graph_A.speration(4)  # shows all cities speratde by distance of 4 
```
