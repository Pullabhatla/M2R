```python
from Req import Map, gen_poly, gen_ran, held_karp
from Req.dynamic import shortest_segment
from time import time
import matplotlib.pyplot as plt
```

### An investigation into the performance of the Held-Karp algorithm

```python
times = []
for i in range(3, 18):
    t0 = time()
    dummy = held_karp(gen_ran(i))
    t1 = time()
    t = t1 - t0
    times.append(t)
    print(f"Time elapsed: {round(t, 3)} seconds")
    dummy.show2d()
```

```python
plt.plot([i for i in range(3,18)], times)
plt.xlabel("Size of map")
plt.ylabel("Evaluation time in seconds")
plt.show()
```
