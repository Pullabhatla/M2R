# Roots of unity

```python
from math import sin, cos, pi
from Req.req import Map
```

## Generating polygons

We shall experiment with the regular polygons generated by the roots of unity in the complex plane and represent our results as a Map in 2D space.

```python
def gen_poly(n, dist_func=None):
    if dist_func:
        return Map([(cos((2*pi*i)/n), sin((2*pi*i)/n)) for i in range(n)], dist_func)
    else:
        return Map([(cos((2*pi*i)/n), sin((2*pi*i)/n)) for i in range(n)])
```

```python
a = gen_poly(4)
l = a.alltours()
for i in l:
    i.show2d()
```
