# Brute force search

```python
from math import sin, cos, pi
from Req import Map, minimal_tour
from time import time
import matplotlib.pyplot as plt
from random import random
```

## Roots of Unity

We shall experiment with the regular polygons generated by the roots of unity in the complex plane and represent our results as a Map in 2D space.

```python
def gen_poly(n, dist_func=None):
    if dist_func:
        return Map([(cos((2*pi*i)/n), sin((2*pi*i)/n)) for i in range(n)], dist_func)
    else:
        return Map([(cos((2*pi*i)/n), sin((2*pi*i)/n)) for i in range(n)])
```

In the program below, we find that the brute force search becomes impractical beyond the octagon.

```python
times = []
for i in range(3, 9):
    t0 = time()
    minimal_tour(gen_poly(i)).show2d()
    t1 = time()
    t = t1 - t0
    times.append(t)
    print(f"Time elapsed: {round(t, 3)} seconds")
```

```python
plt.plot([i for i in range(3,9)], times)
plt.xlabel("Size of map")
plt.ylabel("Evaluation time in seconds")
plt.show()
```

## Random maps

```python
def gen_ran(n, dist_func=None):
    if dist_func:
        return Map([(random(), random()) for i in range(n)], dist_func)
    else:
        return Map([(random(), random()) for i in range(n)])
```

```python
times = []
for i in range(3, 9):
    t0 = time()
    minimal_tour(gen_ran(i)).show2d()
    t1 = time()
    t = t1 - t0
    times.append(t)
    print(f"Time elapsed: {round(t, 3)} seconds")
```

```python
plt.plot([i for i in range(3,9)], times)
plt.xlabel("Size of map")
plt.ylabel("Evaluation time in seconds")
plt.show()
```
