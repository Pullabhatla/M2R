# Brute force search

```python
from Req import Map, brute_tour, gen_poly, gen_ran
from time import time
import matplotlib.pyplot as plt
```

## Roots of Unity

We shall experiment with the regular polygons generated by the roots of unity in the complex plane and represent our results as a Map in 2D space. In the program below, we find that the brute force search becomes impractical beyond the octagon.

```python
times = []
for i in range(3, 9):
    t0 = time()
    dummy = brute_tour(gen_poly(i))
    t1 = time()
    t = t1 - t0
    times.append(t)
    print(f"Time elapsed: {round(t, 3)} seconds")
    dummy.show2d()
```

```python
plt.plot([i for i in range(3,9)], times)
plt.xlabel("Size of map")
plt.ylabel("Evaluation time in seconds")
plt.show()
```
