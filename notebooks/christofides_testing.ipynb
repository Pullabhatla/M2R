```python
from GraphTheory.read import Graph, GraphMatrix
from GraphTheory.mst import MST
from time import time 
import numpy as np 
from GraphTheory.christofides import christofides
from distance_data import grid04, kn57, uk12, wg22, sgb128, wg59
from distance_data import djibouti_adjacencymatrix, usa_adjacencymatrix, ws_adjacencymatrix  
# number of cities respectively 38, 48, 29
kn_57 = kn57.to_numpy()
uk_12 = uk12.to_numpy()
wg_22 = wg22.to_numpy()
sgb_128 = sgb128.to_numpy()
wg_59 = wg59.to_numpy()
data_sets_collected = [uk_12, wg_22, kn_57, wg_59, sgb_128]
data_sets_collected1 = [djibouti_adjacencymatrix, usa_adjacencymatrix, ws_adjacencymatrix]
lis = ['uk_12', 'wg_22', 'kn_57', 'wg_59', 'sgb_128']
lis_cities = [12, 22, 57, 59, 128]
lis1 = ['djibouti',  'usa', 'wsahara']
lis1_cities = [38, 48, 29]
# in data_distances_matrices.ipynb seen that all the adjaceny matrices in 
```

```python
GM = [GraphMatrix(np.triu(i), directed=False) for i in data_sets_collected]  # this the Graph matrix class applied
GM_G = [i.G for i in GM]   # this now allows access to the Graph object versions of the matrices
KGM = [MST(i) for i in GM_G]  # turn into MST objects where kruskals algorithm is a method of the class

GM1 = [GraphMatrix(np.triu(i), directed=False) for i in data_sets_collected1]  # this the Graph matrix class applied
GM_G1 = [i.G for i in GM1]   # this now allows access to the Graph object versions of the matrices
KGM1 = [MST(i) for i in GM_G1]  # turn into MST objects where kruskals algorithm is a method of the class
```

```python
# small detour to kruskals to see run time on all the data sets
kruskal_output = []
times = []
for  i in range(5):
    print(f'kruskal agorithm for {lis[i]} has {str(lis_cities[i])} cities')
    t0 = time()
    b = KGM[i].kruskal()
    kruskal_output.append(b)
    #print(b)
    t1 = time()
    times.append(t1 - t0)
    print(f'timetaken = {t1 - t0}')
kruskal_output1 = []
times1 = []
for  i in range(3):
    print(f'kruskal agorithm for {lis1[i]} has {str(lis1_cities[i])} cities')
    t0 = time()
    b = KGM1[i].kruskal()
    kruskal_output1.append(b)
    #print(b)
    t1 = time()
    times1.append(t1 - t0)
    print(f'timetaken = {t1 - t0}')
```

```python
number_of_cities = lis_cities + lis1_cities
print(number_of_cities)
times_for_all = times + times1
print(times_for_all)
# all times for kruskals min spantree even for 128 nodes
```

```python
list_of_all_cities = lis + lis1
list_of_number_of_cities = lis_cities + lis1_cities
list_of_allmatrices = data_sets_collected + data_sets_collected1
times = []
data_usage = []
for i in range(8):
    t0 = time()
    christofides(list_of_allmatrices[i], name =f'{list_of_all_cities[i]} which has {list_of_number_of_cities[i]} cities')
    t1 = time()
    times.append(t1-t0)
```
