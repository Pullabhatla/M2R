```python
import matplotlib.pyplot as plt 
import numpy as np 
size = [4,5,8,12,15,29,38,131,194,279,380,662,734,929] # from benchmark data
time = [0.04,0.03,0.03,0.05,0.04,0.06,0.07,0.47,1.40,2.05,3.53,27.90, 51.01,86.83] #from bench mark data
p = np.poly1d(np.polyfit(size, time, 3))  # tried degree 3 since PMWM has complexity order 3 
sizes1 = np.linspace(0,1000,50)
times1 =[p(i) for i in sizes1]
plt.plot(size, time, label='benchmark')  #  plot of times from bench markin
plt.plot(sizes1, times1, label='polyfit')  # plot after fitting deg 3 poly to benchmark_data
plt.xlabel("Number of Nodes (N)")
plt.ylabel("Times/s")
plt.legend()
plt.show()
```

```python
print('prediction for 1979:')
print(p(1979))
print(p(1979)/3600)
print('prediction for 3496:')
print(p(3496))
print(p(3496)/3600)
print('prediction for 7146:')
print(p(7146))
print(p(7146)/3600)
print('prediction for 10639:')
print(p(10639))
print(p(10639)/3600)
```

```python
# extrapolating out the times
sizes = np.linspace(0,10000,100)
p = np.poly1d(np.polyfit(size, time, 3))
times =[p(i) for i in sizes]
plt.plot(sizes, times, label='poly fit extrapolation')
plt.plot([1979, 3496], [893, 8157],'o', color='black', label='actual times')
plt.xlabel("Number of Nodes (N)")
plt.ylabel("Times/s")
plt.legend()
plt.show()
```
