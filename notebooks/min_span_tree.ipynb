```python
from GraphTheory.mst import MST
from GraphTheory.read import Graph, GraphMatrix
import numpy as np
import random as rnd
import numpy.random as nrnd
```

```python
# generate a 6 node random graph
rnd.seed(12)  # this means that every time the code is repeated get the same random values
A = nrnd.randint(1,100,[6,6])
print(A)
b =np.array(np.diag(A))
M = A - np.diag(b) # setting the diagonal of the matrix to 0 a self loops is not possible
print(M)
```

```python
D_matrix = GraphMatrix(M)
link_list = D_matrix.link_list
distances = D_matrix.dist_list
print(link_list)
print(distances)
```

```python
D = D_matrix.graph()  # check if it is working
print(D.nodes())
print(D.distances())
print(D.edges())
print(D.edge_dictionary())
print(D.adjacency())
```

```python
KD = MST(D)
print(KD.distances)
print(KD.links)
print(KD.ordered_distance)
print(KD.ordered_links)
```

```python
D.draw()
```

```python
D.draw(edge_weights=True)
```

```python
# kruskla returns set of visited nodes, list of links , distances of these links, total distance.
min_span_tree = Graph(KD.kruskal()[1],KD.kruskal()[2]) 
min_span_tree.draw(directed=False)
```
